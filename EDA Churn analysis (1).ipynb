## Importing Libraries

The project is of Exploratory data analysis. In this data from a telecom company is used. It has main variable of whuy people have churned the service. The data contains many variables such as Gender, Service, Duration and Many others. 
We firstly imported the data and Libraries which we will use. In the Second step we inspect the data and checked if there are any duplicate or null values. After inspecting the data, we changed the data types of categorical columns. We also looked for the descriptive statistics of the data. In the last part, we explored the data. 
Results from the Data are shown visually and numerically in the notebook.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing Data

```python
df= pd.read_csv(r"C:\Users\HP\Downloads\Customer Churn.csv")
df
```

Inspecting the data

```python
df.info()
```

```python
#it shows us the data types and total rows and columns and non null values
```

```python
#the values in column senior citizens have 0 and 1 and we will convert them into yes and no
```

```python
df['SeniorCitizen']= df['SeniorCitizen'].replace({0 : "No", 1: "yes"})
```

```python
df
```

```python
df['SeniorCitizen'].value_counts()
```

```python
#here we change the data type from string to float
```

```python
df["TotalCharges"]= df["TotalCharges"].replace(' ', 0)
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
df.describe()
```

```python
df['PaymentMethod'].value_counts().plot(kind='bar')

plt.title('Count of Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
df['Churn'].value_counts().plot(kind='bar')
plt.title('Count of Churned customers')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
df['Contract'].value_counts().plot(kind='bar')
plt.title('Count of Contract types')
plt.xlabel('Contracts')
plt.ylabel('Count')
plt.show()
```

```python
result = df.groupby(['gender', 'Churn']).size().unstack()
```

```python
result.plot(kind='bar', stacked=True)
```

```python
result
```

```python
plt.figure(figsize= (9,4))
sns.histplot(x= 'tenure', data= df, bins= 72, hue= 'Churn')
plt.show()
```

it shows that most of the churns has been made in the first month or the next months

```python
plt.figure(figsize=(5, 4))
ax= sns.countplot(x= 'Contract', data= df, hue= 'Churn')
ax.bar_label(ax.containers[0])
plt.title('Count of Contract types')
plt.xlabel('Contracts')
plt.ylabel('Count')
plt.show()
```

this figure shows how churn rate is prevaling in contracts and it can be seen that mostly churn rates are from momnth to month contracts

```python
plt.figure(figsize=(5, 4))
ax= sns.countplot(x= 'InternetService', data= df, hue= 'Churn')
ax.bar_label(ax.containers[0])
plt.title('types of services')
plt.xlabel('services')
plt.ylabel('Count')
plt.show()
```

this shows that most churns are from fiber optics services
